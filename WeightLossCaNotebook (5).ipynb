### Removing Outliers 
Below we have a simple example of removing outliers. The Importance of removing outliers from your dataset is there are extreme cases's of data being very distant from the standard distribution. These outliers have the possibility of skewing the results and having an impact on the accuracy of the data.

In this example we're using the method with numpy libary by setting an upper and lower bound of the quantile.

As we can see from the statment our low bound is anything over `1.7` and upper bound anything over `8.9`

```python
lower_bound = 0.02
upper_bound = 0.97
res = weightLoss.wl2.quantile([lower_bound,upper_bound])
res
```

Next we can refined a region to dectect any outliers and we stored this in the value `res` in the code snipnet below anything that is within that region is considered a true index and anything outside of that region is considered a false region

```python
true_index = (res.loc[lower_bound]< weightLoss.wl2.values) & (weightLoss.wl2.values < res.loc[upper_bound])
true_index
```

After we have figured out our outliers we then need to loop through the array and then remove any of those outliers by using the `drop()` method supplied in pandas

```python
i = 0
while i < len(true_index)-1:
    if true_index[i] == False:
        weightLoss.drop(weightLoss.index[i],inplace=True)
    i += 1
weightLoss
```

### Modeling 
Below we are using a powerlifting dataset for our machine learning element
The goal of the machine learning element is to be able to predict the possible first place winner of a powerlifting competition using a decision tree model from *sklearn* 

### Dataset
Just to take note any preprocessing has been done external to this tutorial due to the limited time of the presentation and due to constraints of technical experience in data preprocessing using the supported langauges in Jupter. 

```python
import pandas as pd
powerLifting = pd.read_csv("file:///C:/Users/ahp15/Desktop/Data%20Science/Jupyer%20notebook/powerLifting1.csv", encoding='latin-1', low_memory=False)
powerLifting.head()
```


### Decision Tree
In the below code snipnet we are importing the machine learning model and the graphviz libary to draw the decsion tree. 
We are reading the list of features from the dataset and making the target varrible as `isWinner` and loading the values into a Decision Tree model from SkLearn 

we are define are list of features and target value to make the prediction off of. We then create the model and pass it our labels and features.

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import graphviz 

features = list(powerLifting.columns[1:17])
print("* features:", features, sep="\n")

y = powerLifting["isWinner"]
X = powerLifting[features]
dt = DecisionTreeClassifier(min_samples_split=20, random_state=99)
dt.fit(X, y)
```

### Drawing the tree 
In the below snipnet we are loading in the decision tree created above into the graphviz liabary and open the results of the model in a pdf to visualise it **which should open in another window**

```python
dot_data = tree.export_graphviz(dt, out_file=None, feature_names=X.columns)
graph = graphviz.Source(dot_data) 
graph.render("tree",view = True)
```
